```
#jpg枚数の取得
n_images = !ls -1F | grep -v / | wc -l
n_images = int(n_images[0])
n_images
```

```
#jpgをmovieに変換

import cv2

fourcc = cv2.VideoWriter_fourcc('m','p','4','v')
#cv2.VideoWriter('ファイル名', fourcc, フレームレート, (保存サイズ))
video = cv2.VideoWriter('video.mp4', fourcc, 30, (640, 480))

for i in range(0, n_images):
    img = cv2.imread('masked_picture{0:03d}.jpg'.format(i))
    img = cv2.resize(img, (640,480))
    video.write(img)

video.release()
```
